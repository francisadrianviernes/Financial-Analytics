## ROOT FUNCTION

```python
def f(x):
    return 5*x**3 - np.exp(1)
```

```python
#Initiate our values
high = 1
low = 0

#Calculate the mid and substitute it in our function
mid = (high + low)/2
value = f(mid)
```

```python
convergence = pd.DataFrame(columns=['high', 'mid', 'low'])

iterations = 0
while (value != 0 ) and (abs(value) > 0.000001):
    if value > 0:
        high = mid # assign the midpoint as the new high point
        mid = (high + low)/2
        value = f(mid)
        iterations +=1        
        convergence.loc[iterations] =[high, mid, low]
    else:
        low = mid #for negative values
        mid = (high + low)/2
        value = f(mid)
        iterations +=1
        convergence.loc[iterations] =[high, mid, low]
                        
convergence
```

```python
def bisection(function, high, low, tol=0.01):
    
    '''
    Solves a root of a function using the bisection or binomial search
    method.
    
    Parameters:
    -----------
    function: function object
        - function which root we are solving. Must be a function type
    high: float
        - initial "high" value for the function. Function value must resort
          in an opposite sign as the value from evaluating the function 
          using the "low" value
    low: float
        - initial "low" value for the function. Function value must resort
          in an opposite sign as the value from evaluating the function 
          using the "high" value
    tol: float
        - tolerance for the function to call convergence.
    
    Returns:
    --------
    mid: float
        - final value that will be considered as the (approximate)root
    
    '''
    
    
    #Initiate our values
    high = high
    low = low
    iterations = 0

    #Calculate the mid and substitute it in our function
    mid = (high + low)/2
    value = function(mid)
    
    while (value != 0 ) and (abs(value) > tol):
        if value > 0:
            high = mid # assign the midpoint as the new high point
            mid = (high + low)/2
            value = function(mid)
            iterations +=1    
            
        else:
            low = mid #for negative values
            mid = (high + low)/2
            value = function(mid)
            iterations +=1
    return f'Iteration converged after {iterations} iterations. The final value is {mid}'

bisection(function=f, 
          high=1, 
          low=0, 
          tol=0.01) 
```
