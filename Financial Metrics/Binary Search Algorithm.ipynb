```python
import numpy as np
import pandas as pd
```

```python
def f(x):
    return 15*x**2 - np.exp(x)
```

```python
def binary_search(function, high, low):
    df = pd.DataFrame(columns=["Low", "Mid", "High", "Value of f'(x)"]) # for displaying our results
    high = high
    low = low
    mid = -1 #simply initiatie the variable
    #Check initial case:
    if high > low:
        mid = (high + low)/2
        f_x =  function(mid)
        df.loc[len(df)] = [low, mid, high, round(function(mid),4)]
        while f_x != 0:
            if f_x > 0: #the condition for this variable comes from the high value. That as long as it's positive.
                high = mid
                mid = (high + low)/2
                f_x =  function(mid)
                df.loc[len(df)] = [low, mid, high, round(function(mid),4)]
                display(df)
            else:
                low = mid
                mid = (high + low)/2
                f_x =  function(mid)
                df.loc[len(df)] = [low, mid, high, round(function(mid),4)]
                display(df)
    else:
        return "'High' must be greater than the 'low' parameter"
    return mid
```

```python
binary_search(f, 1,0)
```
