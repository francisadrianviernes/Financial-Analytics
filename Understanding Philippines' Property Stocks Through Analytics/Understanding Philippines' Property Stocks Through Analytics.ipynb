### Crash Course in Stock Beta

Now, let's download some of the stock end price information. Stock beta reveals how sensitive a stock moves with the economy. A beta of one means that a stock moves in perfect correlation with the economy such that if the economy grows by 5%, the stock price is expected to grow by 5%. 


A beta of less than one (1), implies that the stock is less affected by changes in the economy (in this case, we refer to the stock market as a whole, which is used as a gauge for the econmy, albeit less than perfect). If the Philippine stock market grows by 10%, that particular stock grows by less than 10%. The benefit of this is that if the stock market declines by 10%, that stock declines less than that. Oftentimes, we call these stocks as "defensive stocks". They provide good defense in particular, to slow economic growth. Industries such as utility, healthcare and consumer staples belong in this category. 


For stocks that have a beta of greater than one, we refer to these as "cyclical" stocks. As they tend to rise and fall with the Philippine Stock Market as a whole. If the stock market grows by 10%, these stocks appreciate by more than 10%.

The process of calculating beta requires you to get three things: standard deviation of the stock, the standard deviation of the Philippine Stock Exchange and the correlation of their returns.

For the time period: ideally, **we want a time period that encompasses the whole business cycle**. This means a time period that had subperiods of expansion and recession. Ideally, beta captures how a stock return moves with the market in all of the period. 

While professionals have varying preference for this, let's say that 2010, a time where the world is recovering from the global financial crisis, up to 2020, where COVID-19 marked another period of recession, completes a cycle for us. 

#### Getting the beta of individual stocks

````python
#Ensure that the trading date is earlier than current date. You can choose an earlier date 

START_DATE = "2010-10-02"
END_DATE = "2020-10-02"

property_df = pd.DataFrame()
for stock in property_stocks:
    try:
        df_ = get_stock_data(stock, START_DATE, END_DATE)
    except:
        continue
    df_.columns = [stock]
    property_df = pd.concat([property_df, df_], axis=1)
````

````python
merged = psei.merge(property_df, on='dt')
````

Now that we have the merged stock prices for a period of ten years, we can start to process the data and calculate the Beta.

### Formula for Beta

$$Beta = \frac{Cov(r_i, r_m)}{Var(r_m)}$$

where $r_i$ refers to the returns of stock i
and $r_m$ refers to the market returns represented by PSEi returns.

With this, we need to get the individual returns per stock and get their covariance with the market returns.

Luckily, pandas have a built-in method that can simply apply the formula that we have earlier.

````python
#Calculate the returns
ph_stocks_returns = merged.pct_change()
````

````python
#This formulation returns a matrix object (similar to correlation matrix)
beta_df = ph_stocks_returns.cov()/ph_stocks_returns.var()
````

````python
ph_stocks_returns.cov()
````

````python
#unstack would turn this into a series
beta_df = beta_df.unstack()['PSEI'].to_frame()
````

````python
#We reset the index because we plan to label the plot later. Labeling is easier with an integer index.
beta_df = beta_df.reset_index()
beta_df.columns = ['Stock','Beta']
````

### Visualize the results


Now that we have our table of betas, let's visualize and see if we can get insights from the visualization.

````python
plt.figure(figsize=(12,12))
ax = sns.scatterplot(x=beta_df.index, y=beta_df.Beta);

ax.set_title('Beta Values for PSEi Property Stocks',
            fontweight = 'heavy',
            fontsize = 14)

#Drawing a horizontal line where the PSEI lies
ax.axhline(1, c = 'red');

#Text
ax.axes.get_xaxis().set_visible(False);
texts = [ax.text(p[0], p[1]+.005,beta_df.loc[p[0], 'Stock']) for p in zip(beta_df.index, beta_df.Beta)];
````
