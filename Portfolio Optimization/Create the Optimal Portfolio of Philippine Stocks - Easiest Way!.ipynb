```python
#Preliminaries
from fastquant import Portfolio
```

```python
stocks = ['MEG','CEB', 'BDO','ALI', 'MER', 'AC', 'JGS', 'URC', 
          'JFC', 'SEVN', 'BMM', 'CAT', 'FGEN', 'MAXS', 'PGOLD']

p = Portfolio(stocks,"2014-01-01", "2019-12-29")
```

```python
axs = p.data.plot(subplots=True, figsize=(15,10))
fig = p.plot_portfolio(N=10000)
```
