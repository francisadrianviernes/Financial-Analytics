```python
from fastquant import get_stock_data, get_stock_table,  get_pse_data
import pandas as pd
import numpy as np

# Import pyplot from the matplotlib module
from matplotlib import pyplot as plt
```

```python
df = get_pse_data("MEG", "2014-01-01", "2020-07-01")
```

```python
# Plot the daily closing prices
df.close.plot(figsize=(10, 6))


plt.axvline(x='04/05/2015', c='red')
plt.axvline(x='04/08/2019', c='red')

plt.axvline(x='04/30/2020', c='red')


plt.title("Daily Closing Prices of MEG\nfrom 2014-01-01 to 2020-07-01", fontsize=20)
```
